```python
#Initialize the chores and create an empty list to hold the chores
chores = []
def welcome_menu(): #This function will initialize the Menu which will display the List of Available Choices
    print("Welcome to the Stats Division Deliverables!")
    print("1. Add chore")
    print("2. Mark chore as Done")
    print("3. List chores")
    print("4. Remove chore")
    print("5. Quit")

def add_chore(chore_description): # This function will add a chore to the list of chores via way of appending.
    chores.append({"description": chore_description, "done": False})
    print("Chore added!")

def mark_chore_completed(chore_index): # This function is to mark a chore complete if the appropriate entry has been chosen
    if 0 <= chore_index < len(chores):
        if chores[chore_index]["done"]:
            print("Chore is already marked as done.")
        else:
            chores[chore_index]["done"] = True
            print("Chore marked as done.")
    else:
        print("Erroneous chore index.")

def list_chores(): #List Function to list/enumerate all chores added ad their status whether complete or not
    print("chores:")
    for index, chore in enumerate(chores):
        status = "Done" if chore["done"] else "Not Done"
        print(f"{index}. {chore['description']} - {status}")

def remove_chore(chore_index): # This function will pop/remove an item/chore from the list if chore is unneeded 
    if 0 <= chore_index < len(chores):
        removed_chore = chores.pop(chore_index)
        print(f"chore '{removed_chore['description']}' removed.")
    else:
        print("Invalid chore index.")

def main(): # This is the main / overaching function that runs all the other sub-functions in the manager
    while True:
        welcome_menu()
        choice = input("Select an option: ")

        if choice == "1":
            chore_description = input("Enter chore description: ")
            add_chore(chore_description)
        elif choice == "2":
            chore_index = int(input("Enter chore index to mark as done: "))
            mark_chore_completed(chore_index)
        elif choice == "3":
            list_chores()
        elif choice == "4":
            chore_index = int(input("Enter chore index to remove: "))
            remove_chore(chore_index)
        elif choice == "5":
            print("Thanks for using or updating the chore manager. Goodbye and great day!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()

     
```
